# Program # 1:
# Calulator:

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print("Select operation: +  -  *  /  **")
op = input("Enter operator: ")

if op == "+":
    print("Result:", num1 + num2)

elif op == "-":
    print("Result:", num1 - num2)

elif op == "*":
    print("Result:", num1 * num2)

elif op == "/":
    if num2 != 0:
        print("Result:", num1 / num2)
    else:
        print("Division by zero not allowed")

elif op == "**":
    print("Result:", num1 ** num2)

else:
    print("Invalid operator")
```

# Program # 2:
# Check if there is any numeric value in list (Using for loop)

```python
my_list = ["Ameen", "Ali", 25, "Python"]

found = False

for item in my_list:
    if type(item) == int or type(item) == float:
        found = True

if found == True:
    print("Numeric value exists in list")
else:
    print("No numeric value found")
```

# Program # 4:
# Sum all numeric items in dictionary

```python
data = {"a": 10, "b": 20, "c": 30, "d": "Amjad"}

total = 0

for value in data.values():
    if type(value) == int or type(value) == float:
        total = total + value

print("Sum of numeric values:", total)
```

# Program # 5:
# Identify duplicate values from list

```python
numbers = [1, 2, 3, 4, 2, 5, 1, 6]

duplicates = []

for num in numbers:
    if numbers.count(num) > 1:
        if num not in duplicates:
            duplicates.append(num)

print("Duplicate values are:", duplicates)
```
